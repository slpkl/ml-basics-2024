Парсер имел отвратительный код, а данные на выходе были еще хуже. А через пару месяцев я уволился оттуда и дальше проект не полетел (надеюсь, это независимые события). Какой вывод?
````python
def func(input_list: list) -> dict:
    mean_list = (sum([i for i in input_list]))/len(input_list)
    
    numbers_dict = {} 
    for i in input_list:
        value = numbers_dict.get(i, -1)
        if value == -1:
            numbers_dict[i] = input_list.count(i)
    mode_list = sorted(numbers_dict.items(), key=lambda item: item[1])[-1][0] 
    
    list_lenght = len(input_list)
    if list_lenght % 2 == 1: #если чисел в наборе четное количество, то медиана = сред.арифм серединных
        median_list = sorted(input_list)[int(list_lenght/2)]
    elif list_lenght % 2 == 0:
        sorted_input_list = sorted(input_list)
        median_list = (sorted_input_list[int(list_lenght/2)] + sorted_input_list[int(list_lenght/2)-1])/2
    return {"mean": round(mean_list, 4), "median": median_list, "mode": mode_list}
````

````python
def parse_csv(file_content: str) -> list[dict]:
    file_content_lines = file_content.splitlines()
    dict_keys = file_content_lines[0].split(',')
    parsed = []
    length = len(dict_keys)
    for i in range(1, len(file_content_lines)):
        line = file_content_lines[i].split(',')[:length]
        last = file_content_lines[i].split(',')[length-1:]
        str_last = ','.join(last)
        line.append(str_last)
        dict_line = {}
        for j in range (len(dict_keys)-1):
            dict_line[dict_keys[j]] = line[j]  
        dict_line[dict_keys[length-1]] = line[length][1:-1] #в ласт строчке убираем ' ' по бокам из-за замены ' на /' при экранировании   
        parsed.append(dict_line)
    return parsed
````
